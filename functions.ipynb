## Basic function structure

Here's how you create a function:



## Simple Functions
Let's start with functions that don't take any input

```python
# Define the function

def say_hello():
    print("Hello, world!")

# After you run this code, notice there is no output.  You only defined the function.
```

```python
# Call the function

say_hello()
```

## Functions with Parameters
**Parameters** let you pass information into your function

```python
# Defining the functions

def greet_person(name):
    print(f"Hello, {name}!")

def calculate_area(radius):
    area = 3.14159 * radius * radius
    return area
```

```python
# Calling the functions
greet_person("Dwight")

area = calculate_area(5)
print(f"The area is: {area}")   
```

## Default Parameters
You can give parameters default values:

```python
# Defining the function

def greet(name, greeting="Hello"):
    print(f"{greeting}, {name}!")
```

```python
greet("Bob")                    # Uses default: "Hello, Bob!"
greet("Bob", "Hi")
```

## Return Values
Functions can send information back to whoever called them:

```python
# Defining the functions

def square(number):
    return number * number

def get_grade(score):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"
```

```python
# Calling the functions

my_square = square(8)
student_grade = get_grade(85)
print(f"8 squared is {my_square}")
print(f"Grade: {student_grade}")
```

## Real-World Example
Here's a practical example that combines several concepts
**Functions can call other functions!** Here we see the function `print_report_card` calls `calculate_final_grade` and `get_letter_grade`

```python
def calculate_final_grade(homework, tests, participation):
    """Calculate final grade based on weighted averages"""
    final = (homework * 0.3) + (tests * 0.6) + (participation * 0.1)
    return round(final, 1)

def get_letter_grade(numeric_grade):
    """Convert numeric grade to letter grade"""
    if numeric_grade >= 90:
        return "A"
    elif numeric_grade >= 80:
        return "B"
    elif numeric_grade >= 70:
        return "C"
    elif numeric_grade >= 60:
        return "D"
    else:
        return "F"
    
def print_report_card(student_name, homework, tests, participation):
    """Print a complete report card"""
    final_numeric = calculate_final_grade(homework, tests, participation)
    final_letter = get_letter_grade(final_numeric)
    
    print(f"--- Report Card for {student_name} ---")
    print(f"Homework: {homework}%")
    print(f"Tests: {tests}%")
    print(f"Participation: {participation}%")
    print(f"Final Grade: {final_numeric}% ({final_letter})")
```
